## ðŸ§  1. Fancy Indexing kya hota hai?

Fancy indexing me tum directly indices ka list dete ho.

```python
import numpy as np

a=np.array([10,20,30,40,50])

# Select specific elements:

print(a[[0,2,4]])
```

## ðŸ”¥ Fancy indexing = COPY (Important)

```python
b=a[[0,2]]
b[0]=999
print(a)
```

## ðŸŸ¡ 2. Fancy Indexing in 2D

    m[[row_indexes], [col_indexes]]

```python
m = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print(m[[0,2]])

# Select specific elements:

print(m[[0,2],[1,2]])
```

## ðŸ§  3. np.where() â€” ML preprocessing ka king ðŸ‘‘

np.where ka matlab:

ðŸ‘‰ condition true ho to ye karo else wo karo

```python
a=np.array([10,25,30,5,60])

result=np.where(a > 20, "High" ,"Low")
print(result)
```

## Numeric replacement

```python
b=np.where(a>20 , 1,0)
print(b)
```

## np.where in 2D Dataset

```python
x=np.array([
    [1, 50],
    [2, 80],
    [3, 30],
    [4, 90]
])

# Replace low scores with 0
x[:,1] = np.where(x[:,1] < 60, 0, x[:,1])
print(x)
```
