## Slicing (Sub-array)

Syntax:

      array[start : stop : step]

      start = 0            
      stop = stop -1      position


```python
import numpy as np

a =np.array([10,20,30,40,50])

# first element
print(a[0:8])

# till 3 position
print(a[:3])

# Reverse array
print(a[::-1])
```
