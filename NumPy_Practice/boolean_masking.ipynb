## 1-D Masking

```python
import numpy as np

a = np.array([10, 25, 30, 5, 60])
mask = a > 20
print(mask)
print(a[mask])
```

## Boolean Masking in 2D

```python
b = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

mask = b > 50
print(b[mask])
```

## Real ML Use Case: Outliers remove

```python
data = np.array([10, 12, 15, 500, 18])
clean = data[data < 100]
print(clean)
```

## Row Filtering in 2D

```python
data = np.array([
    [1, 50],
    [2, 80],
    [3, 30],
    [4, 90]
])

# print All score
col_score=data[:,1]
print("All Score : ",col_score)

# Filter rows where score > 60
score_60 = data[data[:,1]>60]
print(score_60)
```

# Boolean Masking in 3D

```python
import numpy as np

x = np.array([
    [
        [10, 20],
        [30, 40]
    ],
    [
        [50, 60],
        [70, 80]
    ]
])

print("Shape :",x.shape)

# Condition: values > 50
print(x[x>50])
```

## Real Use Case: Image Processing

```python
img = np.array([
    [50, 200],
    [30, 255]
])

bright=img[img > 100]
print(bright)
```

## Advanced Masking: Replace Values

```python
b = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Replace all values > 50 with 0
b[b > 50]=0
print(b)
```
